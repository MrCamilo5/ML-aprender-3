```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
plt.style.use('fivethirtyeight')
```

```
datos=pd.read_csv('https://www.datos.gov.co/resource/w79x-qwdk.csv')
```

```
datos.head()
```

```
nosirve=[0,2,3,5,6,15]
columnas=datos.columns.values
for i in nosirve:
  datos.drop(columnas[i], axis='columns', inplace=True)
```

```
#Datos promedio por mes
datos2=datos.groupby('mes').mean()
```

```
datos2.plot(subplots=True, figsize=(15,15))
```

```
sns.heatmap(datos2.corr())
```

```
datos2.head()
```

```
sns.heatmap(datos.groupby('departamento').mean().corr(),annot=True)
```

```
X=datos2.iloc[:,[0,1,2,3,5]].values
```

```
y=datos2.iloc[:,[4]].values
```

```
modelo=LinearRegression()
modelo.fit(X,y)
```

```
modelo.predict(X[0].reshape((1,-1))), y[0]
```
